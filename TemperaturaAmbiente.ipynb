Comportamiento de la temperatura en los últimos 60 años en América Central

```python
import pandas as pd
import matplotlib.pyplot as plt

#Importación de CSV
archivo = "Temperatura.csv"
data = pd.read_csv(archivo)

#Validando nombre de columanas
columna_anio = "Años"  
columna_temp = "valor" 
columna_region = "País__ESTANDAR"  


if columna_anio in data.columns and columna_temp in data.columns and columna_region in data.columns:
    #Filtrando America Central
    data_centroamerica = data[data[columna_region] == "América Central"]
    
    if not data_centroamerica.empty:
        #Convirtiendo a numérico
        data_centroamerica[columna_anio] = pd.to_numeric(data_centroamerica[columna_anio], errors="coerce")
        data_centroamerica[columna_temp] = pd.to_numeric(data_centroamerica[columna_temp], errors="coerce")
        
        #Eliminando filas con datos vacíos
        data_centroamerica.dropna(subset=[columna_anio, columna_temp], inplace=True)

        #Configuración de la gráfica
        plt.figure(figsize=(12, 8))

  
        plt.plot(data_centroamerica[columna_anio], data_centroamerica[columna_temp], marker='o', label="América Central")

        #Personalización de la gráfica
        plt.title("Comportamiento de la Temperatura en América Central")
        plt.xlabel("Año")
        plt.ylabel("Temperatura (°C)")
        plt.legend()
        plt.grid(True)


        plt.show()
    else:
        print("No hay datos disponibles para América Central.")
else:
    print("Las columnas necesarias no existen en los datos.")

```

```python
import matplotlib.pyplot as plt

# Configurar la gráfica
plt.figure(figsize=(12, 8))
plt.plot(data_centroamerica[columna_anio], data_centroamerica[columna_temp], marker='o', label="América Central")
plt.title("Comportamiento de la Temperatura en América Central")
plt.xlabel("Año")
plt.ylabel("Temperatura (°C)")
plt.legend()
plt.grid(True)

# Guardar la gráfica en un archivo PNG antes de mostrarla
plt.savefig("Temperatura_Centroamerica.png", format="png", dpi=300)
print("Gráfica guardada como Temperatura_Centroamerica.png")

# Mostrar la gráfica
plt.show()

```

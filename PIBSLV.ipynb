Producto interno bruto (PIB) total anual por habitante a precios constantes en dólares en El Salvador

```python
import pandas as pd
import matplotlib.pyplot as plt

#Importación del CSV
csv_file = 'PibES.csv' 
data = pd.read_csv(csv_file)

#Conversión a numérico
data['Años__ESTANDAR'] = pd.to_numeric(data['Años__ESTANDAR'])

data['value'] = pd.to_numeric(data['value'])

#Configuración de la gráfica
plt.figure(figsize=(12, 6))
plt.plot(data['Años__ESTANDAR'], data['value'], marker='o', label="PIB per cápita")

# Agregar etiquetas y título
plt.title('Producto Interno Bruto (PIB) per cápita en El Salvador (1990-2023)', fontsize=14)
plt.xlabel('Año', fontsize=12)
plt.ylabel('PIB per cápita (dólares)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12)
plt.tight_layout()

plt.savefig("PIBSLV.png", format="png", dpi=300)
print("Gráfica guardada como PIBSLV.png")

# Mostrar la gráfica
plt.show()
```

```python
import seaborn as sns
sns.pairplot(data[['Años__ESTANDAR','value']])
plt.grid(True)

plt.savefig("DispersionPIBSLV.png", format="png", dpi=300)
print("Gráfica guardada como DispersionPIBSLV.png")
```

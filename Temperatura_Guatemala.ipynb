```python
import pandas as pd
import matplotlib.pyplot as plt

#Importaci√≥n del CSV
data = pd.read_csv("CSV_LIMPIOS\Temperatura_Guatemala.csv") 

#Conversi√≥n a num√©rico
data['A√±os__ESTANDAR'] = pd.to_numeric(data['A√±os__ESTANDAR'])
data['value'] = pd.to_numeric(data['value'])

#Creaci√≥n de la gr√°fica
plt.figure(figsize=(10, 6))
plt.plot(data['A√±os__ESTANDAR'], data['value'], marker='o', linestyle='-', color='green', label="¬∞C")

#Personalizaci√≥n
plt.title("Variaci√≥n de temperatura media en Guatemala (1999-2019)")
plt.xlabel("A√±o")
plt.ylabel("Temperatura (¬∞C))")
plt.legend()
plt.grid()

plt.savefig('GRAFICAS/Temperatura_Guatemala.png', dpi=300)
print("Gr√°fica guardada como GRAFICAS/Temperatura_Guatemala.png")


plt.show()
```

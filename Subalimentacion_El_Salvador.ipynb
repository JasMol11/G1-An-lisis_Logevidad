Importaci칩n de CSV CEPALSTAT sobre sub-alimentaci칩n en El Salvador, y gr치fica.

```python
import pandas as pd
import matplotlib.pyplot as plt

#Importaci칩n CSV
archivo_csv = "CSV_SIN_ANALIZAR/Subalimentacion_El_Salvador.csv"  
data = pd.read_csv(archivo_csv)

#Validaci칩n de nombre de las columnas
columna_region = "Pa칤s__ESTANDAR"
columna_bianual = "Bianual"
columna_value = "value"

#Filtraci칩n de datos
data_el_salvador = data[data[columna_region] == "El Salvador"]

#Convirtiendo Bianual a formato num칠rico
data_el_salvador['A침o'] = data_el_salvador[columna_bianual].str[:4].astype(int)

#Configuraci칩n de la gr치fica
plt.figure(figsize=(12, 8))
plt.plot(data_el_salvador['A침o'], data_el_salvador[columna_value], marker='o', label="Subalimentaci칩n en El Salvador")

#Personalizaci칩n
plt.title("Tendencia de Subalimentaci칩n en El Salvador (1999-2019)")
plt.xlabel("A침o")
plt.ylabel("N칰mero de personas en subalimentaci칩n (en miles)")
plt.legend()
plt.grid(True)

#Guardando imagen como png
plt.savefig("GRAFICAS/Subalimentacion_El_Salvador.png", format="png", dpi=300)
print("Gr치fica guardada como GRAFICAS/Subalimentacion_El_Salvador.png")

#Mostrando la gr치fica
plt.show()
```

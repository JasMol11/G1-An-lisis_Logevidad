```python
import pandas as pd
import matplotlib.pyplot as plt

# Importación del CSV
data = pd.read_csv("CSV_LIMPIOS/Temperatura_El_Salvador.csv")

# Conversión de columnas relevantes a tipo numérico
data['Años__ESTANDAR'] = pd.to_numeric(data['Años__ESTANDAR'], errors='coerce')
data['value'] = pd.to_numeric(data['value'], errors='coerce')

# Eliminar filas con valores nulos
data = data.dropna(subset=['Años__ESTANDAR', 'value'])

# Eliminar valores duplicados en la columna Años__ESTANDAR
data = data.drop_duplicates(subset=['Años__ESTANDAR'])

# Ordenar los datos por la columna de los años
data = data.sort_values(by='Años__ESTANDAR')

# Creación de la gráfica
plt.figure(figsize=(10, 6))
plt.plot(data['Años__ESTANDAR'], data['value'], marker='o', linestyle='-', color='green', label="°C")

# Personalización de la gráfica
plt.title("Variación de temperatura media en El Salvador (1961-2022)")
plt.xlabel("Año")
plt.ylabel("Temperatura (°C)")
plt.legend()
plt.grid()

# Guardar la gráfica como imagen
plt.savefig('GRAFICAS/Temperatura_El_Salvador.png', dpi=300)
print("Gráfica guardada como GRAFICAS/Temperatura_El_Salvador.png")

# Mostrar la gráfica
plt.show()
```

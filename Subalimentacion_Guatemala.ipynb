```python
import pandas as pd
import matplotlib.pyplot as plt

# Importar el archivo CSV
file_path = 'CSV_LIMPIOS/Subalimentacion_Guatemala.csv'  
df_subalimentacion_guatemala = pd.read_csv(file_path)  # Asignamos el nombre al DataFrame

# Revisar los datos
print(df_subalimentacion_guatemala.head())

# Configurar el gráfico
plt.figure(figsize=(10, 6))
plt.plot(df_subalimentacion_guatemala['Bianual'], df_subalimentacion_guatemala['value'], 
         marker='o', linestyle='-', color='red', label='Personas en subalimentación')

# Personalizar la gráfica
plt.title('Número de personas en subalimentación en Guatemala (1999-2019)')
plt.xlabel('Periodo Bianual')
plt.ylabel('Número de personas (miles)')
plt.xticks(rotation=45)
plt.grid(visible=True, linestyle='--', alpha=0.7)
plt.legend()

#Guardando imagen como png
plt.savefig("GRAFICAS/Subalimentacion_Guatemala.png", format="png", dpi=300)
print("Gráfica guardada como GRAFICAS/Subalimentacion_Guatemala.png")

# Mostrar la gráfica
plt.tight_layout()
plt.show()
```

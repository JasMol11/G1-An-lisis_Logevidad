Producto interno bruto (PIB) total anual por habitante a precios constantes en dólares en Guatemala

```python
import pandas as pd
import matplotlib.pyplot as plt

#Importación del CSV
data = pd.read_csv("PibGM.csv") 

#Conversión a numérico
data['Años__ESTANDAR'] = pd.to_numeric(data['Años__ESTANDAR'])
data['value'] = pd.to_numeric(data['value'])

#Creación de la gráfica
plt.figure(figsize=(10, 6))
plt.plot(data['Años__ESTANDAR'], data['value'], marker='o', linestyle='-', color='green', label="PIB per cápita")

#Personalización
plt.title("Producto Interno Bruto (PIB) per cápita en Guatemala (1990-2023)")
plt.xlabel("Año")
plt.ylabel("PIB per cápita (dólares)")
plt.legend()
plt.grid()

plt.savefig("PIBG.png", format="png", dpi=300)
print("Gráfica guardada como pibg.png")


plt.show()

```

```python
import seaborn as sns
sns.pairplot(data[['Años__ESTANDAR','value']])
plt.grid(True)

plt.savefig("DispersiónG.png", format="png", dpi=300)
print("Gráfica guardada como DispersiónG.png")
```
